# Understanding Density Histograms

This notebook demonstrates how to create and interpret **density histograms** using `matplotlib`. Unlike a standard histogram which shows frequency counts, a density histogram is normalized such that the total area of all its bars equals 1. This makes the y-axis represent a **probability density**.

```python
# Use the anline magic command to display plots directly in the notebook output.
%matplotlib inline
# Import the NumPy library for numerical operations.
import numpy as np
# Import the pyplot module from Matplotlib for creating plots.
import matplotlib.pyplot as plt
# Import the function to load the wine dataset from scikit-learn.
from sklearn.datasets import load_wine
```

## 1. Simple Density Histogram (Equal Bins)

We start with a small dataset and create a histogram where the data range is divided into 5 bins of equal width. By setting `density=True`, the y-axis is scaled to represent probability density, not the raw count of items in each bin.

```python
# Define a small dataset of marks.
marks = [35, 36, 37, 42, 48, 53, 58, 70, 73, 84]

# Create a histogram with 5 equal-width bins.
# density=True normalizes the histogram so the area of the bars sums to 1.
plt.hist(marks, bins=5, density=True, edgecolor='red')

# Add a title and labels to the plot for clarity.
plt.title("Density Estimation: Marks (Equal Bins)")
plt.xlabel("Marks")
plt.ylabel("Density")

# Display the plot.
plt.show()
```

## 2. Density Histogram (Unequal Bins)

Histograms can also have bins of unequal widths. In this case, using `density=True` is crucial. It ensures that the **area** of each bar (height Ã— width) is proportional to the number of data points it contains, providing a fair visual representation of the data's distribution.

```python
# Define custom, unequal bin edges.
bin_edges = [35, 42, 45, 54, 60, 85]
print("Custom Bin Edges:", bin_edges)

# Calculate the width of each custom bin.
Bin_widths = np.diff(bin_edges)
print("Width of each bin:", Bin_widths)

# Create a histogram using the custom bin edges.
# The height of each bar is adjusted for its width to maintain the density property.
plt.hist(marks, bins=bin_edges, density=True, edgecolor='red')

plt.title("Density Estimation: Marks (Unequal Bins)")
plt.xlabel("Marks")
plt.ylabel("Density")
plt.show()
```

## 3. Example with a Large, Normally Distributed Dataset

Let's see how histograms look with a large dataset generated from a standard normal distribution.

```python
# Create a modern random number generator for reproducibility.
rng = np.random.default_rng(1701)

# Generate 10,000 random samples from a standard normal distribution (mean=0, std=1).
data = rng.normal(size=10000)
```

### Frequency Histogram

First, a standard histogram where the y-axis shows the **frequency (count)** of samples in each bin.

```python
plt.hist(data)
plt.title("Frequency Histogram of Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show();
```

### Density Histogram

Now, a density histogram of the same data. The shape is identical, but the y-axis is now scaled to represent **probability density**.

```python
plt.hist(data, bins=20, density=True)
plt.title("Density Histogram of Normal Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show();
```

## 4. Example with Real-World Data (Wine Dataset)

Let's apply this to a real dataset from scikit-learn to visualize the distribution of alcohol content in wine samples.

```python
# Load the dataset.
wine = load_wine()

# Extract the 'Alcohol' feature.
alcohol = wine.data[:, 0]
```

```python
# Create a density histogram to visualize the distribution of alcohol content.
plt.hist(alcohol, bins=20, density=True)
plt.title("Density Estimation: Alcohol Content")
plt.xlabel("Alcohol")
plt.ylabel("Density")
plt.show()
```

## 5. Styling with `rwidth`

The `rwidth` parameter controls the relative width of the bars. A value of 1.0 makes the bars touch, while a value less than 1.0 (like 0.8) will create gaps. Values greater than 1.0 cause the bars to overlap, which is usually not desired.

```python
# Setting rwidth=1 makes the bars touch each other, a common style.
plt.hist(alcohol, rwidth=1, density=True, edgecolor='black')
plt.title("Density Estimation (rwidth=1)")
plt.xlabel("Alcohol")
plt.ylabel("Density")
plt.show()
```
