```python
# Basic imports
import os
os.environ["KERAS_BACKEND"] = "tensorflow"
import numpy as np
import tensorflow as tf
import keras
from keras import layers
import matplotlib.pyplot as plt

### 1. DATA LOADING AND PREPROCESSING ###

# Load the Fashion-MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Combine train and test sets for unsupervised learning
fmnist_images = np.concatenate([x_train, x_test], axis=0)

# Preprocess the data:
# 1. Add a channel dimension for grayscale images (28, 28) -> (28, 28, 1)
# 2. Convert to float32 and normalize pixel values to the range [0, 1]
fmnist_images = np.expand_dims(fmnist_images, -1).astype("float32") / 255.0
print(f"Dataset shape: {fmnist_images.shape}")


### 2. VAE MODEL COMPONENTS ###

# A custom layer to sample from the latent space distribution
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding an image."""
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch_size = tf.shape(z_mean)[0]
        latent_dim = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch_size, latent_dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# Define the Encoder network
latent_dim = 10 # Dimensionality of the latent space

encoder_inputs = keras.Input(shape=(28, 28, 1))
x = layers.Conv2D(32, 3, activation="relu", strides=2, padding="same")(encoder_inputs)
x = layers.Conv2D(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Flatten()(x)
x = layers.Dense(16, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
print("\nEncoder Summary:")
encoder.summary()

# Define the Decoder network
latent_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(7 * 7 * 64, activation="relu")(latent_inputs)
x = layers.Reshape((7, 7, 64))(x)
x = layers.Conv2DTranspose(64, 3, activation="relu", strides=2, padding="same")(x)
x = layers.Conv2DTranspose(32, 3, activation="relu", strides=2, padding="same")(x)
# Use a sigmoid activation to ensure output values are in the [0, 1] range
decoder_outputs = layers.Conv2DTranspose(1, 3, activation="sigmoid", padding="same")(x)
decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
print("\nDecoder Summary:")
decoder.summary()


### 3. VAE MODEL CLASS WITH CUSTOM TRAINING STEP ###

class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        # Metrics to track the loss components
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        # List the metrics to be displayed during training
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            # Get the latent space parameters from the encoder
            z_mean, z_log_var, z = self.encoder(data)
            # Reconstruct the image from the sampled latent vector
            reconstruction = self.decoder(z)
            
            # --- Calculate the two-part loss ---
            # 1. Reconstruction Loss (how well we recreated the input)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction),
                    axis=(1, 2), # Sum over the image pixels
                )
            )
            # 2. KL Divergence Loss (a regularizer for the latent space)
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            
            # Total Loss is the sum of both components
            total_loss = reconstruction_loss + kl_loss
            
        # Compute and apply gradients
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        
        # Update and log the metrics
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        
        return {m.name: m.result() for m in self.metrics}


### 4. COMPILE AND TRAIN THE VAE ###

# Instantiate the VAE model
vae = VAE(encoder, decoder)

# Compile the model with an optimizer
vae.compile(optimizer="adam")

# Train the model
print("\nStarting VAE training...")
history = vae.fit(
    fmnist_images,
    epochs=25, # For a good result, 25-30 epochs are recommended
    batch_size=128
)
print("Training complete.")


### 5. GENERATE AND VISUALIZE SYNTHETIC IMAGES ###

def plot_generated_images(decoder_model, n=10, latent_dim=10, figsize=(10, 10)):
    """Generates and plots a grid of synthetic images."""
    # Create a grid of n x n images
    plt.figure(figsize=figsize)
    # Sample random points from the standard normal distribution
    random_latent_vectors = tf.random.normal(shape=(n * n, latent_dim))
    # Decode them to create synthetic images
    generated_images = decoder_model.predict(random_latent_vectors, verbose=0)
    
    for i in range(n * n):
        # Get the i-th image and remove the channel dimension
        img = generated_images[i].reshape((28, 28))
        # Add a subplot to the grid
        ax = plt.subplot(n, n, i + 1)
        plt.imshow(img, cmap="gray")
        plt.axis("off")
    
    plt.suptitle(f"{n*n} Synthetically Generated Fashion-MNIST Images", fontsize=16)
    plt.show()

print("\nGenerating and displaying synthetic images...")
# 
plot_generated_images(decoder, n=10, latent_dim=latent_dim)
```
